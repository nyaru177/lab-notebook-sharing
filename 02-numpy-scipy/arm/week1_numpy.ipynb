```python
# 练习3：实现 softmax（数值稳定版）
def softmax(x):
    max = np.max(x,axis=-1,keepdims=True)
    demaxval = x-max
    xexp = np.exp(demaxval)
    xsum = np.sum(xexp,axis=-1,keepdims=True)
    xsoftmax = xexp / xsum
    return xsoftmax
```

```python
import numpy as np
# 练习4：实现 one-hot 编码
def one_hot(labels, num_classes):
    N = len(labels)
    oh = np.zeros((N,num_classes),dtype=np.float32)

    oh[np.arange(N),labels] = 1.0
    # labels: shape (N,)，整数标签
    # 返回: shape (N, num_classes) 的 float32 矩阵
    return oh
```

```python
# 练习5：向量化实现 k-NN（最近邻搜索）
def knn_distances(query, database):
    # query: (Q, D)
    # database: (N, D)
    Q, D = query.shape
    N, _ = database.shape

    # 每一个query向量L2平方和 (Q,1)
    q_sq = np.sum(query ** 2, axis=1, keepdims=True)
    # 每一个database向量L2平方和 (1,N)
    d_sq = np.sum(database ** 2, axis=1, keepdims=True).T
    # 点积 (Q,N)
    dot = query @ database.T

    # 欧式距离平方
    dist_sq = q_sq - 2 * dot + d_sq
    # 防止浮点极小负数，截断到0
    dist_sq = np.maximum(dist_sq, 0.0)
    # 开根号得到标准欧式距离
    dist = np.sqrt(dist_sq)

    return dist
```
