## 1. Markdown Cell：让实验记录可读

训练中的交叉熵可以写成行内公式 $L = -\sum_i y_i \log(\hat{y}_i)$。

注意力计算也可以写成块级公式：

$$
\operatorname{Attention}(Q, K, V) = \operatorname{softmax}\left(\frac{QK^T}{\sqrt{d_k}}\right)V
$$

| 内容 | 适合使用的 Cell |
| --- | --- |
| 可执行的 Python 代码 | Code |
| 实验目的、解释和结论 | Markdown |
| 需要原样保留的文本 | Raw |

```python
import sys

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

print('Python:', sys.version.split()[0])
print('NumPy:', np.__version__)
```

## 2. Code Cell：运行代码并保留结果

先用一个小数组完成计算，再把结果画出来。真实实验中，数据加载、模型定义、训练和评估也应按这种顺序分成多个 Cell。

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(6, 3.5))
ax.plot(x, y, label='sin(x)')
ax.set_title('A small reproducible experiment')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.legend()
ax.grid(alpha=0.25)
fig.tight_layout()
plt.show()
```

## 3. Magic Command：记录一个简单的运行时间

`%timeit` 适合快速比较小段代码的运行时间。它是 Jupyter 的交互功能，不是普通 Python 脚本语法。

```python
matrix = np.arange(100).reshape(10, 10)
%timeit -n 3 -r 1 matrix @ matrix.T
```
