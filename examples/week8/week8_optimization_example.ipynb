## 学习目标

1. 在固定数据、指标和随机种子下比较 baseline 与改动。
2. 用图表和数值说明优化是否有效。

> 这是一个最小的优化示例，只演示“相同数据、指标和随机种子下比较 baseline 与改动”。完整项目还需要加入数据增强、模型结构或训练策略优化中的至少两项。


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score

rng = np.random.default_rng(42)
y_true = rng.integers(0, 2, size=300)
scores = np.clip(0.25 + 0.5 * y_true + rng.normal(0, 0.2, 300), 0, 1)

def evaluate(threshold):
    y_pred = (scores >= threshold).astype(int)
    return f1_score(y_true, y_pred)

baseline_threshold = 0.5
baseline_f1 = evaluate(baseline_threshold)
thresholds = np.linspace(0.1, 0.9, 81)
f1_values = np.array([evaluate(t) for t in thresholds])
best_index = f1_values.argmax()
best_threshold = thresholds[best_index]
best_f1 = f1_values[best_index]

print(f'baseline: threshold={baseline_threshold:.2f}, F1={baseline_f1:.4f}')
print(f'optimized: threshold={best_threshold:.2f}, F1={best_f1:.4f}')
print(f'improvement: {best_f1 - baseline_f1:+.4f}')
```

```python
plt.figure(figsize=(7, 4))
plt.plot(thresholds, f1_values, label='Validation F1')
plt.axvline(baseline_threshold, color='gray', linestyle='--', label='Baseline')
plt.axvline(best_threshold, color='tab:red', linestyle=':', label='Best threshold')
plt.xlabel('Decision threshold')
plt.ylabel('F1')
plt.title('A reproducible optimization comparison')
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
