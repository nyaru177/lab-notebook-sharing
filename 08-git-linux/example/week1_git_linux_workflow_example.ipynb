## 1. Git 提交流程

推荐的单个 Notebook 提交流程如下。命令中的 `your-name`、`topic` 和路径需要替换成自己的内容。

```bash
git switch main
git pull
git switch -c add-your-name-topic
git add 04-sklearn/your-name/week2_topic_your-name.ipynb
git commit -m "Add topic notebook"
git push -u origin add-your-name-topic
```

流程含义：先同步 `main`，再在个人分支上提交；推送后在 GitHub 创建 Pull Request，由仓库维护者检查并合并。若小组明确允许直接提交 `main`，也应保留同样的命名、运行和检查要求。

````python
import platform
import shutil
import subprocess
import sys


def command_version(command):
    """跨平台检查命令是否可用，不执行修改文件的操作。"""
    executable = shutil.which(command)
    if executable is None:
        return f'{command}: not found'
    result = subprocess.run(
        [executable, '--version'],
        capture_output=True,
        text=True,
        check=False,
    )
    output = (result.stdout or result.stderr).strip().splitlines()
    return output[0] if output else f'{command}: available'


print('OS:', platform.system())
print('Python:', sys.version.split()[0])
print(command_version('git'))
print(command_version('conda'))
````

## 2. Linux 常用命令对照

| 目的 | Linux / macOS | PowerShell |
| --- | --- | --- |
| 查看当前位置 | `pwd` | `Get-Location` |
| 查看目录内容 | `ls` | `Get-ChildItem` |
| 进入目录 | `cd path` | `Set-Location path` |
| 复制文件 | `cp a b` | `Copy-Item a b` |
| 移动或重命名 | `mv a b` | `Move-Item a b` |
| 搜索文本 | `grep pattern file` | `Select-String pattern file` |
| 查找文件 | `find . -name '*.ipynb'` | `Get-ChildItem -Recurse -Filter '*.ipynb'` |

在 Notebook 中应把这些命令写在 Markdown 说明里，并注明操作系统；不要依赖某台电脑上的绝对路径。
